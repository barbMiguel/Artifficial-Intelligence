```python
#####################################
# Autor: Miguel Angel Osorio Sanchez
# Autor: Juan Camilo Cubillos Ayala
# Universidad Icesi - Cali, Colombia
#####################################

#####################################
# Tic-tac-toe
#####################################

####### SETUP #######

import sys
global ia_play

MIN = -1
MAX = 1

EASY = 0
HARD = 1

#States in which the game is over
winner_states = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5,8], [0, 4, 8], [2, 4, 6]]


####### Definitions #######
def winner(board):   
    win = 0
    for state in winner_states:
        if board[state[0]] == board[state[1]] and board[state[0]] == board[state[2]] and board[state[0]] != 0:
            win = board[state[0]]
    return win


def game_over(board):
    no_board = False
    for i in range(0, len(board)):
        if board[i] == 0:
            no_board = True
            
    if winner(board) == 0 and no_board:
        return False
    else:
        return True
    
####### Min-Max application #######
# Computer's turn
def minmax(board, player, level):
    global ia_play
    
    #Game over
    if game_over(board):
        return [winner(board),0]
    plays = []
    
    for play in range(0,len(board)):
        
        if board[play] == 0:
            temp_board = board[:]
            temp_board[play] = player
            score = minmax(temp_board,player*(-1),level)
            plays.append([score,play])

    if level == EASY:
        if player == MAX:
            move = min(plays,key=str)
            ia_play=move[1]
            return move
        else:      
            move = max(plays, key=str)        
            return move[0]
    if level == HARD:
        if player == MAX:
            move = max(plays, key=str)        
            ia_play = move[1]
            return move 
        else:       
            move = min(plays, key=str)        
            return move[0]

####### Computer's turn #######
def ia_turn(board, level):
    global ia_play
    punt = minmax(board[:], MAX, level)
    board[ia_play] = MAX
    return board


####### Player's turn #######
def player_turn(board):
    finished= False
    while not finished:
        inp = input("Type the number of box (from 1 to 9), type exit or do not type for exit: ")
        print("Typed %s " % (inp) )
        
        if inp == '' or inp == 'exit':
            sys.exit()
        if len(inp) == 1:
            box = int(inp)
            if board[box-1]==0:
                board[box-1]=MIN
                finished=True                
            else:
                print("please type a non played box")
        else:
            print("please type only one number")
    return board
```

```python
########################
####### THE GAME #######
########################

def print_board(board):
    
    current_board = list(map(str, board))    
    for i in range(0, len(board)):
        if board[i] == MAX:
            current_board[i] = 'X'
        elif board[i] == MIN:
            current_board[i] = 'O'
        else:
            current_board[i] = ' '
  
    print('   |   |')
    print(' ' + current_board[0] + ' | ' + current_board[1] + ' | ' + current_board[2])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + current_board[3] + ' | ' + current_board[4] + ' | ' + current_board[5])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + current_board[6] + ' | ' + current_board[7] + ' | ' + current_board[8])
    print('   |   |')
    
####### DIFFICULTY LEVEL #######

def select_level(board):
    level= False
    while not level:
        selected_level = input("Please type level: 0 for easy and 1 for hard")
        if selected_level in '01'  and len(selected_level)==1:
            level = True
        else:
            print("please type a valid level")
    selected_level = int(selected_level)
    return selected_level
```

```python
####### START #######
def start_game():
    print("Starting game... player starts")
    board = [0,0,0,0,0,0,0,0,0]
    level = select_level(board)
    
    if level == EASY:
        print("Easy mode")        
    if level == HARD:
        print("Hard mode")

    while True:        
        print_board(board)
        board = player_turn(board)
        if game_over(board):
            break
        board = ia_turn(board, level)
        print("Is game over?")
        if game_over(board):
            break
            
    print_board(board)
    win = winner(board)
    if win == 0:
        win = "Tie"
    elif win == MIN:
        win = "Player"
    else:
        win = "Computer"
    
    print("Winner: %s" % (win))
    
start_game()
```
